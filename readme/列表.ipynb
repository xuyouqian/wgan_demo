# python 列表

## 索引

按照索引可以把列表里的元素取出来

### 正向索引

正向索引从0开始，到列表的长度减1 结束

```python
L = ['A','B','C']
L[0]
```

```python
L[1]
```

### 反向索引

```python
反向索引 从-1 开始 - 列表的长度结束
```

```python
L[-1]
```

```python
L[-3]
```

## 切片

取出列表的多个元素，要输入起始索引，终止索引，步长,取出的元素从起始索引开始，不包含终止索引，间隔距离为步长

```python
L = list('ABCDEFGH')
L
```

```python
L[0:4]  # 步长默认为1
```

```python
L[0:4:2]
```

```python
L[3::-1] # 起始索引和终止都可以省略，省略标识取到所有
```

## 索引赋值

根据索引修改列表里的元素

```python
L = list('ABCDEFGH')
L
```

```python
L[-1] = 'Z'
L
```

## 切片赋值

```python
L = list('ABCDEFGH')
L
```

```python
L[2:5:1] = ["P","Y","P"]
L 
```

```python
L[2:5:1] = ["a","b","c","d"]
L 
```

```python
# 切片可以凭空给列表添加元素
L = list('ABCDEFGH')
L
```

```python
# 在 B C 中间插入一个数字 1
L[2:2] = [1]
L 
```

```python
# 在 B 1 中间插入两个字符
L[2:2] = ['x','y']
L 
```

## 扩展斐波那契列表

```python
L = [1,1,2]
```

```python
# 取出列表里的最后一个元素
first_num = L[-1]
# 取出列表里倒数第二个元素
second_num = L[-2]
# 下一个斐波那契数
next_num = first_num + second_num 
# 把下一个斐波那契数丢到列表里面

L.append(next_num)
L 
```

```python
L = [1,1,2]
for i in range(3,100):
    # 取出列表里的最后一个元素
    first_num = L[-1]
    # 取出列表里倒数第二个元素
    second_num = L[-2]
    # 下一个斐波那契数
    next_num = first_num + second_num 
    # 把下一个斐波那契数丢到列表里面

    L.append(next_num)

L 
# len(L)
```

```python
L = [1,1,2]

while L[-1] <= 100:
    # 取出列表里的最后一个元素
    first_num = L[-1]
    # 取出列表里倒数第二个元素
    second_num = L[-2]
    # 下一个斐波那契数
    next_num = first_num + second_num 
    # 把下一个斐波那契数丢到列表里面

    L.append(next_num)
    

# L = L[:-1]
L.pop(-1)
L 
```
